```python
# 8 bit pix2pix
import os
import numpy as np
from PIL import Image

# Load the data
mnist_x = np.load("/home/alingold/Training/20230714_minlap_700crop_ref_pro.npy")
mnist_b = np.load('/home/alingold/Training/20230714_minlap_700crop_ccm_pro.npy')

# Create the pix2pixData directory and the test, val, and train subdirectories
base_dir = '/home/alingold/DP-GAN/datasets/20230714_minlap_700crop_pro_pix2pix'
os.makedirs(base_dir, exist_ok=True)

sub_dirs = ['test', 'val', 'train']
for sub_dir in sub_dirs:
    os.makedirs(os.path.join(base_dir, sub_dir), exist_ok=True)

# Define the number of images for each category
test_size = 100
val_size = 100
train_size = len(mnist_x) - test_size - val_size

# Divide the dataset into categories
for idx, (image_x, image_b) in enumerate(zip(mnist_x, mnist_b)):
    # Concatenate the images horizontally
    combined_image = np.concatenate((image_x, image_b), axis=1)

    # Scale the pixel values to the 0-255 range
    combined_image = (combined_image * 255).astype(np.uint8)

    # Convert the combined image to a PIL Image object
    combined_image_pil = Image.fromarray(combined_image, mode='L')

    # Determine the category and save the combined image
    if idx < test_size:
        category = 'test'
    elif idx < test_size + val_size:
        category = 'val'
    else:
        category = 'train'

    # Save the combined image as n.jpg inside the corresponding subdirectory
    # Add 1 to the index to start naming the files from 1.jpg
    file_path = os.path.join(base_dir, category, f'{idx + 1}.jpg')
    combined_image_pil.save(file_path)
```

```python
#makes black or white squares for blurry and non-blurry images

import os
import numpy as np
from PIL import Image
import random

mnist_x = np.load("20230914_not_blurry.npy")
mnist_b = np.load('20230914_blurry.npy')

# Create directory structure
base_dir = '/home/alingold/pix2pix/datasets/20230914_blurrynotblurry_pix2pix'
os.makedirs(base_dir, exist_ok=True)

for sub_dir in ['test', 'val', 'train']:
    os.makedirs(os.path.join(base_dir, sub_dir), exist_ok=True)

test_size = 30
val_size = 0
train_size = min(len(mnist_x), len(mnist_b)) - test_size - val_size

white_square = np.ones((256, 256), dtype=np.uint16)
black_square = np.zeros((256, 256), dtype=np.uint16)

idx_x = 0
idx_b = 0
idx = 0

while idx_x < len(mnist_x) or idx_b < len(mnist_b):
    rand_choice = random.randint(0, 1)

    if rand_choice == 1 and idx_x < len(mnist_x):
        image_x = mnist_x[idx_x, :, :]
        idx_x += 1
        combined_image = np.concatenate((image_x, white_square), axis=1)
    elif rand_choice == 0 and idx_b < len(mnist_b):
        image_b = mnist_b[idx_b, :, :]
        idx_b += 1
        combined_image = np.concatenate((image_b, black_square), axis=1)
    else:
        continue

    combined_image = (combined_image * 65535).astype(np.uint16)
    combined_image_pil = Image.fromarray(combined_image, mode='I;16')

    if idx < test_size:
        category = 'test'
    elif idx < test_size + val_size:
        category = 'val'
    else:
        category = 'train'

    file_path = os.path.join(base_dir, category, f'{idx + 1}.png')
    combined_image_pil.save(file_path)
    idx += 1
```

```python
import os
import numpy as np
from PIL import Image
# dataset augmentation
# Load the data
mnist_x = np.load("/home/alingold/Training/20230714_minlap_700crop_ref_pro.npy")
mnist_b = np.load('/home/alingold/Training/20230714_minlap_700crop_ccm_pro.npy')

# Create the pix2pixData directory and the test, val, and train subdirectories
base_dir = '/home/alingold/pix2pix/datasets/20230714_minlap_700crop_pro_aug20_pix2pix'
os.makedirs(base_dir, exist_ok=True)

sub_dirs = ['test', 'val', 'train']
for sub_dir in sub_dirs:
    os.makedirs(os.path.join(base_dir, sub_dir), exist_ok=True)

total_images = len(mnist_x)

# Get indices to flip vertically and horizontally
flip_horizontal_indices = np.random.choice(total_images, int(total_images * 0.2), replace=False)
flip_vertical_indices = np.random.choice(total_images, int(total_images * 0.2), replace=False)

# Define the number of images for each category
test_size = 100
val_size = 100
train_size = total_images - test_size - val_size

# Divide the dataset into categories
for idx, (image_x, image_b) in enumerate(zip(mnist_x, mnist_b)):
    
    # Flip images horizontally
    if idx in flip_horizontal_indices:
        image_x = np.fliplr(image_x)
        image_b = np.fliplr(image_b)
        
    # Flip images vertically
    if idx in flip_vertical_indices:
        image_x = np.flipud(image_x)
        image_b = np.flipud(image_b)

        # Store the flipped images back into the arrays
    mnist_x[idx] = image_x
    mnist_b[idx] = image_b

# Divide the dataset into categories
for idx, (image_x, image_b) in enumerate(zip(mnist_x, mnist_b)):
    # Concatenate the images horizontally
    combined_image = np.concatenate((image_x, image_b), axis=1)

    # Scale the pixel values to the 16-bit range (0-65535)
    combined_image = (combined_image * 65535).astype(np.uint16)

    # Convert the combined image to a PIL Image object
    combined_image_pil = Image.fromarray(combined_image, mode='I;16')

    # Determine the category and save the combined image
    if idx < test_size:
        category = 'test'
    elif idx < test_size + val_size:
        category = 'val'
    else:
        category = 'train'

    # Save the combined image as n.png inside the corresponding subdirectory
    # Add 1 to the index to start naming the files from 1.png
    file_path = os.path.join(base_dir, category, f'{idx + 1}.png')
    combined_image_pil.save(file_path)
```
